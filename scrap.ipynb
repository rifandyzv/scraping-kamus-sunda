```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
hulu = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

page = r"https://id.wiktionary.org/wiki/Lampiran:Kamus_bahasa_Sunda_%E2%80%93_bahasa_Indonesia"
pageTree = requests.get(page, headers=hulu)
page = soup(pageTree.content, 'html.parser')
```

```python
dataKecap = page.find_all("ul")
```

```python
kamus = {}


for daptar in dataKecap :
    daptarKecap = daptar.find_all("li")
    for kecap in daptarKecap :
        try :
            sunda = kecap.select('a')[0].get_text(strip=True)
            indonesia = kecap.select('a')[1].get_text(strip=True)
            if (len(sunda)) == 5 :
                kamus[sunda] = indonesia
        except Exception :
            pass
        # cokotKecap = kecap.find_all("a")
        # print(cokotKecap)
        # try :
        #     dapetKecap = cokotKecap.text
        #     if (len(dapetKecap)) == 5 :
        #         print(dapetKecap)
        # except Exception :
        #     # lajeng
        #     pass
            
kamus

        
    

        
```
